```python
import os
```

```python
my_dir="/content/drive/MyDrive/Images for Facial Recognition"
#intro to os.walk
print("start print")
for root_dir_path,sub_dirs,files in os.walk(my_dir):
  print("Root Directory Path",root_dir_path)
  print("Sub Directories:",sub_dirs)
  print("Files",files)
  
  print("end print")
```

```python
#walking a directory tree-print only files
```

```python
my_dir="/content/drive/MyDrive/Images for Facial Recognition"
for dir,sub,files in os.walk(my_dir):
  print("Print Dir: ",dir)  #print out each directory path
  sub=[n for n in sub] #subs and files
  contents=sub+files #subs and files
  contents.sort() #sort subs and files

  for f in contents:
    if os.path.isfile(f):
      print("\tJust The Files",f)
  print()
      #print spaces between levels
```

```python
#Printing only sub-folders 
#Print only subs

for dir,sub,files in os.walk(my_dir):
  print("Print Dir:",dir)
  sub=[n for n in sub]
  sub.sort()

  for f in sub:
    print('\tJust The Subs',f)
  print()
```

```python
#count number of files in dirs and sub dirs
count_file=0
dir1="/content/drive/MyDrive/Images for Facial Recognition"
for path,sub,filenames in os.walk(dir1):
  sub=[n for n in sub]
  content=sub+filenames
  for f in content:
    count_file+=1
    print("count: ",count_file,":",f)
print("Final count is:",count_file)
```

```python
#Print Directories Selectively(1)
#print paths that don't starts with "_"

for path,sub_dirs,files in os.walk(my_dir):
  if not path.endswith("_"):
    print("Print path:",path)
```

```python
#Print Directory Sekectively(2)
#Print all directories except those that starts with number:"_" (i.e "1.")

for root,dirs,files in os.walk(dir1,topdown=True):
  for dir in dirs:
    if not dir[1]==".":
      print(dir)
```
